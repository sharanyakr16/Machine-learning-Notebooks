```python
X=np.linspace(0,10,30).reshape(30,1)#shape(30,) nothing ao reshaping it as 30,1
y=X+ 2*np.random.normal(size=X.shape)#adding noise  to X and saying both the vectors have the same shape

theta = np.zeros((1,X.shape[1]+1))
```

```python
#linear model
def h_theta(X,theta):
    #append 1 for bias term
    X-bias = np.append(X,np.ones((X.shape[0],1)),axis=1)#making [X,1] and then can do dot product of Q0,Q1 and X,1
    #calculate ax+b
    yhat = np.dot(X_bias,theta.T)#T-transpose of theta
    return yhat
```

```python
#mean square error
def MSE(y,X,theta,h_theta):
    m=y.size  # no of points we are taking averages at tjose many points
    yhat=h_theta(X,theta) #calculate corresponding values of current hypothesis function at those points
    error = np.sum((yhat-y)**2)/2*m 
    return error
```
